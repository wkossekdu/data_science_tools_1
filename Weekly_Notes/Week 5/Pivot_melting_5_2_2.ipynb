```python
import pandas as pd
import seaborn as sns

titanic_df=sns.load_dataset('titanic')
titanic_df.head()
```

```python
p_titanic = titanic_df.drop_duplicates('age').pivot(index='age', columns = 'class', values = 'fare')
p_titanic.head(15)
```

```python
pt_titanic = titanic_df.pivot_table(index='age', columns='class', values='fare')
pt_titanic.tail()
```

```python
# FutureWarning: The provided callable <function min at 0x0000028743D1D300> is currently using DataFrameGroupBy.min. 
# In a future version of pandas, the provided callable will be used directly. To keep current behavior pass the string "min" instead.


import numpy as np
pt_titanic = titanic_df.pivot_table(index='age', columns="class", values='fare', aggfunc=np.min)
pt_titanic.tail()
```

```python
print(p_titanic.head())
titanic_copy = p_titanic.copy()
titanic_copy.columns = titanic_copy.columns.astype(str)
melted_titanic_df = pd.melt(
    titanic_copy.reindex(),
    #id_vars=['age'],
    var_name='class_renamed',
    value_vars=['First','Second','Third'],
    value_name='fare'
)
melted_titanic_df.sample(5)
```
