## Demo on the gutenberg text

Let us go to: https://www.gutenberg.org. 

Find a story and list most frequently used words.

Use curl url, tr to lower case, grep -oE "\w+", sort, uniq -c, sort -nr and head

```python
# Let us work on the above commands

!curl -O https://www.gutenberg.org/files/59761/59761-0.txt
```

```python
!curl -s https://www.gutenberg.org/files/59761/59761-0.txt
```

```python
!curl https://www.gutenberg.org/files/59761/59761-0.txt
```

```python
# grep will not work on Windows, but it should on Linux
!curl https://www.gutenberg.org/files/59761/59761-0.txt | grep -oE "\w+"
```

```python
!curl https://www.gutenberg.org/files/59761/59761-0.txt | grep -oE "\w+" | sort | uniq -c 
```

```python
!curl https://www.gutenberg.org/files/59761/59761-0.txt | grep -oE "\w+" | sort | uniq -c | sort -nr |head -n 10
```
