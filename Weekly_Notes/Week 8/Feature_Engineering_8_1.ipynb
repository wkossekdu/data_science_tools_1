# Lecture 8.1. Feature Engineering

## What is feature engineering

Using expert/domain knowledge to create new features from the data. Objective of feature engineering is to create efficient and more generalizable algorithms. 

Converting data to relevant numerical features requires good understanding of domain knowledge.

In the real world this is rarely the case.

We need to convert

- Categorical data
- Text data
- Image data

to numerical and tabular format.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib
titanic_df = pd.read_csv('https://raw.githubusercontent.com/utsav507/titanic-data-analysis/master/titanic_train.csv')
```

# Categorical data encoding using sklearn

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit_transform(titanic_df.Embarked.dropna())
#titanic_df.Embarked.dropna().transform(le.fit_transform)
```

```python
titanic_df.dtypes
```

```python
titanic_df.Embarked = titanic_df.Embarked.astype('category')
```

```python
titanic_df['Embarked_code']=titanic_df.Embarked.cat.codes
titanic_df.head()
```

```python
titanic_df[titanic_df.Embarked.isnull()]
```

# Use one-hot encoding to denote presence or absence of a category

```python
from sklearn import preprocessing
titanic_df = pd.read_csv('https://raw.githubusercontent.com/utsav507/titanic-data-analysis/master/titanic_train.csv')
le = preprocessing.OneHotEncoder(sparse=False)
le.fit_transform(titanic_df.Embarked.dropna().values.reshape((-1,1)))
```

# Categorical data encoding using pandas

We can do this directly in pandas. Pandas supports this feature using get_dummies.

```python
pd.get_dummies(titanic_df, columns=['Embarked'])
```

# Custom encoding (optional)

Let us say we only care whether you embarked from S, C or not

```python
titanic_df['is_S_C_embarked'] = np.where(titanic_df.Embarked.str.contains("S|C"),1,0)
```
